### indexing and slicing in numpy

```python
'''
 - indexing and sclicing in numpy array is the same as the indexing and slicing as in other datatypes
 - BUT ==> in numpy it doesnot take a copy of the array its accessing the actual array
 
 - ex
 - b = a[1:5]
 - b --> part of array (a) , accessing the same index of the array
'''
```

```python
A = np.arange(100)
b = A[3:10]
```

```python
b
```

```python
b[0] = -1200
```

```python
A
```

```python
# if want to have a copy of the datatype

b = A[1:10].copy()
b
```

```python
A[3] = 3
```

```python
A
```

```python
A[::5] # printing all elements step = 5
```

```python
A[::-5] # steps is from the end
```

```python
A[::-1]
```

```python
A[59] = 888
```

```python
# finding the index of an element in the array

count = 0

for i in A: 
    if i == 888:
        print(count)
    count+=1
```

```python
# make it in a function

def index_array(array , element):
    count = 0
    for i in array:
        if i == element:
            print(count)
        count+=1
```
